# Control Flow in Python: If-Else, For Loop, While Loop

### 1. If-Else Statements
If-Else is used to make decisions in code based on conditions.

**Syntax:**
```python
if condition:
    # code block
elif another_condition:
    # code block
else:
    # code block
```

````python
height = float(input('Enter Your height: '))
age = float(input('Enter Your age: '))
if height < 5:
    print("You can't Ride")
elif height > 6.5:
    print("You also can't ride")
else:
    print('You can ride')

if height >= 5 and height <= 6.5:
    if age > 18:
        print('Go Ride')
    else:
        print('Get Out')
````

### 2. For Loops
For loops are used to iterate over a sequence (like a list, tuple, string, or range).

**Syntax:**
```python
for variable in range(start, stop, step):
    # code block
```

````python
for x in range(0, 10, 2):  # from 0 to 9, step of 2
    print(x)
````

### 3. While Loops
While loops repeat as long as a condition is `True`.

**Syntax:**
```python
while condition:
    # code block
    break  # optional, to manually stop the loop
```

````python
condition = True
while condition:
    password = input('Passwords: ')
    if password == 'stop':
        break
````
