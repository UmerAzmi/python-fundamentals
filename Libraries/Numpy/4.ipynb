
# 3. NumPy in AI QR Code Project

In your **AI QR Code Generator and Decoder project**, you have used NumPy mainly for:
1. Handling image data (pixels).
2. Normalizing values for models.
3. Reshaping arrays into the right format.
4. Converting data into PyTorch tensors.

This notebook explains those parts step by step with examples.


```

import numpy as np
from PIL import Image

# Open an image (example placeholder, replace with your file if testing)
img = Image.new("RGB", (3,3), color=(100,150,200))  
arr = np.array(img)
print(arr)
print("Shape:", arr.shape)
```


### Image to NumPy Array
- We open an image using **PIL**.
- Convert it to a NumPy array with `np.array(img)`.
- Each pixel is stored as **[R, G, B]** (Red, Green, Blue values).  
- The `.shape` tells us:
  - Height (rows)
  - Width (columns)
  - Channels (3 for RGB)

Example: `(3, 3, 3)` → 3 rows, 3 cols, 3 color channels.


```

# Normalize values (0–255) → (0–1)
arr_norm = arr / 255.0
print(arr_norm)
print("Min:", arr_norm.min(), "Max:", arr_norm.max())
```


### Normalization
- Images have pixel values from **0 to 255**.
- Deep learning models work better with small, consistent values.  
- So we scale values between **0 and 1** by dividing by 255.

Now `arr_norm` is ready for input into a model.


```

# Reshape example (batch_size, height, width, channels)
arr_reshaped = arr_norm.reshape(1, arr.shape[0], arr.shape[1], arr.shape[2])
print(arr_reshaped.shape)
```


### Reshaping
- Models expect data in specific shapes.  
- Example: `(1, 3, 3, 3)` → `1 image`, 3×3 size, 3 channels.

Here we added **batch size = 1** in the front.


```

import torch

# Convert NumPy array to PyTorch tensor
tensor_img = torch.tensor(arr_reshaped, dtype=torch.float32)
print(tensor_img.shape)
print(type(tensor_img))
```


### Converting to PyTorch Tensor
- Models like Stable Diffusion or ControlNet need **tensors** (PyTorch format).
- `torch.tensor()` converts from NumPy → Tensor.  
- Shape stays the same, but the type changes to `torch.Tensor`.

Now the image is fully ready for deep learning processing.
