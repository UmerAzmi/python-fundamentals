
# Decorators and Context Managers in Python

## 1. Decorators

A **decorator** is a special function that can modify the behavior of another function without changing its code.  
We usually use the `@decorator_name` syntax.

Example:  
- A decorator can run extra code **before** and **after** a function call.
- A decorator function takes a function as a argument


```python
# Simple decorator example
def my_decorator(func):
    def wrapper():
        print('Good Morning')
        func()
        print('World')
    return wrapper

@my_decorator
def say_goodmorning():
    print("Hello")

say_goodmorning()
```


### Example with an imported decorator

Sometimes decorators can be defined in a separate module and imported.  
Here we assume `myModule.py` defines a decorator called `mydec`.


```python

# Example (this will work only if myModule.py exists)
from myModule import mydec

@mydec
def say_hello():
    print("Sir")

# Call function (decorator modifies its behavior)
say_hello()
```


## 2. Context Managers

A **context manager** is used to properly manage resources, like files.  
It ensures resources are cleaned up (like closing a file) automatically.

We usually use context managers with the `with` keyword.


```python

# Context Manager with file
with open("testfile.txt", "r") as f:
    content = f.read()
print(content)
```


### Example: Custom Context Manager with `__enter__` and `__exit__`


```python

class my_context:
    def __enter__(self):
        print('My Name')
        return 'Slim Shaddy'
    def __exit__(self, exc_type, exc_val, exc_tb):
        print('My Name again')

with my_context() as t:
    print(t)
```


### Example: Custom File Manager Context

We can create our own file manager class using `__enter__` and `__exit__`.


```python

class FileManager:
    def __init__(self, filename, mode):
        self.filename = filename
        self.mode = mode
        self.file = None

    def __enter__(self):
        print("Opening file...")
        self.file = open(self.filename, self.mode)
        return self

    def __exit__(self, exc_type, exc_value, traceback):
        print("Closing file...")
        self.file.close()

# Using the context manager
with FileManager("testfile1.txt", "w+") as f:
    f.file.write("Hey my G, we just used a custom context manager!\n")
    print("Writing complete.")
    f.file.seek(0)
    print("Reading full file:\n", f.file.read())
```
