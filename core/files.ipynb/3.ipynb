
# Data Structures in Python

In Python, we use different data structures to store and organize data:

1. **List** → Ordered, mutable (can change), allows duplicates.  
2. **Tuple** → Ordered, immutable (cannot change), allows duplicates.  
3. **Set** → Unordered, mutable, no duplicates allowed.  
4. **Dictionary** → Key-Value pairs, mutable, keys must be unique.  


## Lists (`[]`)

```python
# Lists are ordered collections that can store multiple values.
# They allow duplicates and can hold mixed data types.
# We can add, remove, or update items in a list (mutable).

fruits = ['apple','mango',4]  # FRUITS VARIABLE
print('Lists:\n','1:-',fruits[0],' , 2:-',fruits[1],' , 3:-',fruits[2])

# Insert and modify
fruits.insert(2,'strawberry')
print(fruits)

fruits[3] = 'blueberry'
print(fruits)
```

```python
# Remove items
fruits.remove('blueberry')
print(fruits)

# Extend with multiple items
fruits.extend(['blackberry','blueberry','watermelon'])
print(fruits,'\nIteration by Items')

# Common operations
fruits.append('watermelon')  # Add new item at the end
fruits.pop(1)                # Remove item by index
fruits.insert(0, 'Apple')    # Insert at a specific index
print("Updated List:", fruits)
```

```python
# Iteration by items
for fruit in fruits:
    print(fruit)

print('\nIteration By Item in Fruits:')
for fruit in fruits:
    if fruit == 'blueberry':
        print('OMG')
    else:
        print('NMG')

print('\nIteration By Items in Range:')
for x in fruits:
    if x[-1] == 'y':
        print('Ends with y')
    else:
        print('NO')

# List comprehension
li=[1,2,3,4,5,6,7,8,9]
def func(x):
    return x**x

print('\nList comprehension (even numbers squared by themselves): ',[func(x) for x in li if x%2==0])

# Slice operator
print('\nSlice Operations:\n',fruits[2:4])
```

## Tuples (`()`)

```python

# Tuples are ordered collections like lists, but they are immutable (cannot be changed).
# Useful for fixed data (like coordinates, days of the week, etc.).

vegetable = ('potato','tomato', 2)
print(f'\nTuple: {vegetable}\n', '1:-', vegetable[0], ' , 2:-', vegetable[1], ' , 3:-', vegetable[2])

# Tuple methods
print(f'Count "tomato": {vegetable.count("tomato")}')
print(f'Index of "tomato": {vegetable.index("tomato")}')
```

## Sets (`{}`)

```python

# Sets are unordered collections of unique elements.
# No duplicates allowed.
# Used when you want to store unique items only.

id = {1, 4, 2, 1, 5}  # Duplicate '1' will be removed automatically
print(f'\nSets: {id}')

# Set operations
id.add(3)       # Adds 3
id.add(1)       # Adding duplicate has no effect
id.remove(2)    # Removes 2 (error if not present)
id.discard(8)   # Removes 8 if present, no error if not
print(f'Updated Sets: {id}')
```

## Dictionaries (`{key: value}`)

```python

# Dictionaries store data in key-value pairs.
# Keys must be unique and immutable (like strings, numbers).
# Values can be any type (mutable).

students = {
    101: {'name': 'Umer', 'age': 22, 'grade': 'A', 'city': 'Mumbai'},
    102: {'name': 'Azmi', 'age': 23, 'grade': 'B', 'city': 'Mumbai'},
    103: {'name': 'Z', 'age': 23, 'grade': 'A+', 'city': 'Delhi'}
}

hundred = {'name': 'Umer', 'age': 22, 'grade': 'A', 'city': 'Mumbai'}

# Accessing dictionary data
print(f'\nDictionary: {students}')
print('Name of 103:', students[103]['name'])
print('Age of 101:', students[101]['age'])
print('Single dictionary:', hundred['name'], '\n')

# Dictionary methods
print("All keys:", students.keys())
print("Values of student 102:", students[102].values())
print("Items of student 102:", students[102].items())
```

## Collections.Counter

```python

from collections import Counter

# Counter is a special dictionary in Python (from collections module).
# It helps us count how many times each element appears in a list, string, or any iterable.

# 1: Counting characters in a string
print("\nCounting characters in a string")

s = "supercalifragilisticexpialidocious"
count = Counter(s)
print("Counter:", count)

# Accessing frequency of a specific character
print("Frequency of 'e':", count['e'])
print("Frequency of 'x':", count['x'])
```

```python
# 2: Counting words in a list
print("\nCounting words in a list")

words = ["apple", "banana", "apple", "cherry", "banana", "apple"]
word_count = Counter(words)
print("Word Counter:", word_count)

print("Frequency of 'apple':", word_count['apple'])
print("Frequency of 'cherry':", word_count['cherry'])

# 3: Iterating over Counter
print("\nIterating over Counter")
for item, freq in count.items():
    print(f"Character '{item}' appears {freq} times.")

# 4: Most Common Elements
print("\nMost Common Elements")
print("Most common element(s):", count.most_common(1))
print("Top 2 common elements:", count.most_common(2))
```

## String Methods

```python

# String methods are useful for manipulating and analyzing text data.

text = '  Umer Azmi.47 '  # TEXT VARIABLE
print("\nString method:\n" + text.strip())  # removes all the empty spaces or gaps
print(len(text))     # Return the length of the string
print(text.upper())  # Uppercase
print(text.lower())  # Lowercase
print(text.split())  # Splits the string into a list

# .find() and .count()
print('Finds z on: ',text.find('z'))
print('Count of m: ',text.count('m'))

# Slice Operator [start:stop:step]
print('\nSlice Operations:\n' + text[2:6])
print(text[:6])
print(text[2:])
print(text[2::2])
```
