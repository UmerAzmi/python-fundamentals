# 📊 Pandas Basics Tutorial
This notebook covers the basics of using **Pandas** for data analysis.
You'll learn how to:
- Load data
- Explore data
- Filter, update, and clean data
- Perform grouping, sorting, and exporting

Useful for quick **revision**.

```python
import pandas as pd
```

## 📥 Reading a CSV File

```python
df = pd.read_csv(r'C:\\Users\\Umer Azmi\\Downloads\\products.csv')
print('Print Function:')
print(df)

# `display()` is better inside Jupyter because it shows data in a table format
display(df)
```

## 👀 Viewing Data

```python
# Shows first 5 rows
display(df.head())

# Shows last 5 rows
df.tail()
```

## 📝 Dataset Info & Summary

```python
display(df.info())  # General info about dataset
df.describe()  # Statistical summary for numeric columns
```

## 🔎 Columns and Index

```python
display(df.columns)  # List of all columns
display(df.index)    # Index range of rows
```

## 🎯 Selecting Columns

```python
display(df['Name'].head())   # Single column = Series
print(f'Total Entries: {len(df["Name"])}')

display(df[['Name','Brand']])   # Multiple columns = DataFrame
```

## 📌 Selecting Rows with `iloc`

```python
display(df.iloc[0])   # First row

display(df.iloc[0][0])   # First row, first column
df.iloc[0]['Stock']      # Specific column in first row
```

## 🔍 Filtering Data

```python
display(df[df['Category'] == "Men's Clothing"])   # Simple filter

display(df[(df['Price'] >= 500) & (df['Availability'] == 'in_stock')])  # AND condition

display(df[(df['Color'] == 'Black') | (df['Category'] == "Men's Clothing")])  # OR condition
```

## ✨ String Functions in Filters

```python
display(df[df['Category'].str.lower().str.startswith('laptop')])
display(df[df['Category'].str.lower().str.endswith('clothing')])
df[df['Category'].str.contains('clothing', case=False)]
```

## 📋 Checking Membership with `isin`

```python
display(df[df['Size'].str.lower().isin(['l', 'large'])])
display(df[~df['Size'].str.lower().isin(['l', 'large'])])  # ~ = NOT
```

## ✏️ Updating Values

```python
df.loc[df['Name'] == 'Tablet', 'Size'] = 'M'
display(df[df['Name'] == 'Tablet'])
```

## ❌ Dropping Rows

```python
df = df.drop(99)  # Drops row at index 99
df.tail()
```

## 🚿 Handling Missing Values (NaN)

```python
df = df.dropna()  # Removes rows with missing values

df.fillna({'Size': 'One Size Fits All'}, inplace=True)  # Fill missing values
```

## 🏷️ Renaming Columns

```python
df.rename(columns={'Index': 'SR Number'}, inplace=True)
```

## 🔢 Value Counts

```python
df['Size'].value_counts()
```

## 📊 Grouping & Sorting

```python
display(df.groupby('Availability')['Price'].sum())  # Group by Availability

df.sort_values('Price', ascending=False).head(10)   # Top 10 expensive items
```

## 💾 Saving to CSV

```python
df.to_csv('newfile.csv', index=False)  # Save changes to a new CSV file
```
