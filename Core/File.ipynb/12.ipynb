
# String Formatting (Tables) and Ternary Expressions in Python

In this notebook, we will learn about:
1. **Printing headers in a table** using string formatting.
2. **Ternary expressions** (short form of if-else).



## 1. Printing Headers in a Table

- We can display data in a nice table format using **f-strings with width specifiers**.  
- Example: `f"{value:10}"` → reserves **10 spaces** for that value.

Let's look at an example:


````python

print("Example 1: Printing headers in a table\n")

# Printing the table header
print(f"{'Name':15} {'Age':5} {'City':15}")
print("-" * 40)  # separator line

# Printing some rows with fixed column widths
print(f"{'Alice':15} {25:<5} {'New York':15}")
print(f"{'Bob':15} {30:<5} {'London':15}")
print(f"{'Charlie':15} {22:<5} {'Paris':15}")
print(f"{input('Enter a name'):15} {input('Enter Age'):<5} {input('Enter the City'):15}")
````


### Explanation:
- `{'Name':15}` → "Name" takes up 15 spaces (left aligned by default).  
- `{25:<5}` → number `25` takes up 5 spaces, aligned to the **left** (`<`).  
- `{'City':15}` → "City" takes up 15 spaces.  
- `"-" * 40` → prints 40 dashes as a separator line.



## 2. Ternary Expressions

- A **ternary expression** is a shortcut for `if-else` in one line.  
- Syntax:  
```python
value_if_true if condition else value_if_false
```

Let's see an example:


````python

print("Example 2: Ternary Expressions\n")

age = 18

# Normal if-else
if age >= 18:
    status = "Adult"
else:
    status = "Minor"

print("Normal if-else:", status)

# Using ternary expression (short form)
status = "Adult" if age >= 18 else "Minor"
print("Using ternary expression:", status)
````


### Another example with booleans


````python

book = {"title": "1984", "borrowed": True}

# Normal if-else
if book["borrowed"]:
    status = "Borrowed"
else:
    status = "Available"

print(f"Book: {book['title']} → {status}")

# Ternary expression
status = "Borrowed" if book["borrowed"] else "Available"
print(f"Book (ternary): {book['title']} → {status}")
````


### Explanation:
```python
"Borrowed" if book["borrowed"] else "Available"
```
- If `book["borrowed"]` is `True` → status = `"Borrowed"`  
- If `book["borrowed"]` is `False` → status = `"Available"`  

👉 This is a concise and readable way to write simple if-else conditions.
