
# Overriding Methods in Python

In this notebook, we will learn about **method overriding** and **operator overloading** in Python using a `point` class example.



## 1. What is Method Overriding?

- **Overriding methods** means redefining built-in methods or parent class methods inside your own class.  
- In Python, we can override special methods (also called **dunder methods**) like `__add__`, `__sub__`, `__str__`, etc.  
- This lets us change how operators (like `+`, `-`, `*`, `/`) work with our own objects.

We will create a `point` class to demonstrate this.


```python

class point:
    def __init__(self,x,y):
        self.x = x
        self.y = y

    def coorinate(self):
        return (self.x,self.y)

    # Operator Overloading
    def __add__(self,other):
        return point(self.x + other.x, self.y + other.y)
    def __sub__(self,other):
        return point(self.x - other.x, self.y - other.y)
    def __mul__(self,other):
        return self.x * other.x + self.y * other.y
    def __truediv__(self,other):
        return self.x / other.x + self.y / other.y
    def __floordiv__(self,other):
        return self.x // other.x + self.y // other.y
    def __mod__(self,other):
        return self.x % other.x + self.y % other.y

    # String representation
    def __str__(self):
        return f'Point({self.x},{self.y})'

    # Move method
    def move(self,x,y):
        self.x += x
        self.y += y

    # Length method
    def length(self):
        return (self.x**2 + self.y**2)**0.5

    # Comparison Overloading
    def __gt__(self,other):
        return self.length() > other.length()
    def __ge__(self,other):
        return self.length() >= other.length()
    def __le__(self,other):
        return self.length() <= other.length()
    def __lt__(self,other):
        return self.length() < other.length()
    def __eq__(self,other):
        return self.length() == other.length()
    def __ne__(self,other):
        return self.length() != other.length()
```


## 2. Using the `point` Class

Now let's create objects and test the overridden operators and methods.


```python

p1 = point(1,2)
p2 = point(3,4)
print('P1:',p1,'\nP2:',p2)

# Arithmetic Operations
p3 = p1 + p2
p4 = p1 - p2
print('Addition:',p3)
print('Subtract:',p4)
print('Multiply:',p1 * p2)
print('Divide:',p1 / p2)
print('Int Divide:',p1 // p2)
print('Modulus:',p1 % p2)

# Moving points
p1.move(1,2)
print('New P1:',p1)
p2.move(-1,-2)
print('New P2:',p2)

# Comparisons
print('Greater than:',p1 > p2)
print('Greater than Equal to:',p1 >= p2)
print('Less than:',p1 < p2)
print('Less than Equal to:',p1 <= p2)
print('Equal to:',p1 == p2)
print('Not Equal to:',p1 != p2)

# Coordinates
print('P2 coordinates:', p2.coorinate())
```
