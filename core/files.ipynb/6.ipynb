
# Classes and Inheritance in Python

In this notebook, we will learn about **classes** and **inheritance** in Python with simple explanations and examples.



## 1. What is a Class?

A **class** is like a blueprint for creating objects.  
Objects created from a class can have:
- **Attributes** (variables that belong to the object)
- **Methods** (functions that belong to the object)

Let's look at an example of a class called `dog`.


```python

class dog(object):
    def __init__(self,name,age):
        self.name = name
        self.age = age
        self.favNum= [4,7,11]

    def speak(self):
        print('Hi my name is', self.name, ', I am', self.age, 
              'years old and my favourite numbers are', self.favNum)

    def change_age(self,age):
        self.age = age

    def add_weight(self,weight):
        self.weight = weight

    def talk(self):
        print('Bark!')
```


## 2. Creating and Using Objects

We can now create objects of the `dog` class and use its methods.


```python

print('Dogs:')
asim = dog('Asim',30)
sallo = dog('Sallo',20)

asim.speak()
sallo.speak()

# Changing age and adding weight
sallo.change_age(21)
sallo.add_weight('50Kg')
print('Asim Says:',end=' ')
asim.talk()

sallo.speak()
print('My weight is', sallo.weight)
```


## 3. Inheritance

**Inheritance** allows one class (child) to use the properties and methods of another class (parent).  
This makes it easier to reuse code.

Let's make a `cat` class that inherits from the `dog` class.  
We will also override the `talk` method to make cats say "Meow!" instead of "Bark!".


```python
class cat(dog):
    def __init__(self,name,age,colour):
        super().__init__(name,age)
        self.colour = colour
    def talk(self):
        print('Meow!')

print('\nCats:')
faiya = cat('Faiya',30,'blue')
frha = cat('Frha',30,'red')

faiya.speak()
frha.speak()
faiya.talk()
```


## 4. Another Example of Inheritance: Vehicles

We can also make a more practical example with vehicles.  
We create a base class `vehicle` and then create child classes `car` and `truck` that inherit from it.


```python

class vehicle(object):
    def __init__(self,colour,price,gas):
        self.colour = colour
        self.price = price
        self.gas = gas

    def fillUpTank(self):
        self.gas = 100
    def emptyTank(self):
        self.gas = 0
    def gasLeft(self):
        return self.gas

class car(vehicle):
    def __init__(self,colour,price,gas,speed):
        super().__init__(colour,price,gas)
        self.speed = speed
    def horn(self):
        print('beep beep')

class truck(vehicle):
    def __init__(self,colour,price,gas,tires):
        super().__init__(colour,price,gas)
        self.tires = tires
    def horn(self):
        print('Honk Honk')

# Testing the classes
print('\t\n Inheritance Example')
a = vehicle('black',800,90)
a.fillUpTank()
print(a.gasLeft())

b = car('red',900,20,1000)
b.emptyTank()
b.horn()
```
