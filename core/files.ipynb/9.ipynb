
# Map, Filter, and Lambda Functions in Python

In this notebook, we will learn about:
- **Map function**
- **Filter function**
- **Lambda functions (anonymous functions)**



## 1. Map Function

- The `map()` function applies a function to each element in a sequence (like a list).  
- Syntax: `map(function, iterable)`  
- Returns a map object (we usually convert it to a list).

Let's see an example:


```

li = [1,2,3,4,5,6,7,8,9]

def func(x):
    return x**x

# Without Map
newlist = []
for x in li:
    newlist.append(func(x))
print('Without Map:', newlist)

# With Map
print('With Map:', list(map(func, li)))
```


## 2. Filter Function

- The `filter()` function filters elements of a list based on a condition.  
- Syntax: `filter(function, iterable)`  
- Returns only the elements where the function returns `True`.

Example: Filter out odd numbers.


```

def isodd(x):
    return x % 2 != 0

a = list(filter(isodd, li))
b = list(map(func, a))

print('Filter Result (A):', a)
print('After applying func on A (B):', b)
```


## 3. Lambda Functions

- A **lambda function** is an anonymous (nameless) function defined using the keyword `lambda`.  
- Syntax: `lambda arguments: expression`  

Example:


```

print('Normal Function:', isodd(5))

isOdd = lambda x: x % 2 != 0
print('Lambda Function:', isOdd(5))

# Using lambda with map
print('Map with Lambda function:', list(map(lambda x: x**x, li)))
```
