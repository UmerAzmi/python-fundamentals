✅ **Practice Task:**
1. Create a list of 10 numbers and print the even numbers only.
2. Write code to find the third largest element in a list.
3. Reverse a list without using slicing (`[::-1]`).

```python
nums = [1,5,3,7,9,5,9,2,12,51]
n=4
```

```python
# 1. Create a list of 10 numbers and print the even numbers only.
s =[]
for num in nums:
    if num%2==0:
        s.append(num)
print(s)

# Using List Comprehension
a = [num for num in nums if num %2==0]
print(a)

# Using Lambda and Filter Function
print(list(filter(lambda x: x%2==0, nums)))
```

```python
# 2. Write code to find the third largest element in a list.
First=Second=Third=float('-inf')
for num in nums:
    if num > First:
        Third = Second
        Second = First
        First = num
    elif num > Second and num != First:
        Third = Second
        Second = num
    elif num > Third and num != First and num != Second:
        Third = num

print(f'Largest Element: {First} \nSecond Largest Element: {Second} \nThird Largest Element: {Third}')

# Using Sets
num = list(set(nums))
print(f'\nLargest Element: {num[-1]} \nSecond Largest Element: {num[-2]} \nThird Largest Element: {num[-3]}')
```

```python
# 3. Reverse a list without using slicing (`[::-1]`).
a = []
for num in nums:
    a.insert(0,num)
print(a)
```

✅ **Practice Task:**
1. Take a sentence as input and count the number of vowels.
2. Write a program to check if two strings are anagrams.

```python
# 1. Take a sentence as input and count the number of vowels.
sentence = input('Enter a sentence: ')
print('The Sentences is: ',sentence)
count = 0
for i in sentence.lower():
    if i in 'aeiou':
        count += 1
print(count)

# Using Generator Expression
print('\nThe Sentences is: ',sentence)
count = sum(1 for i in sentence.lower() if i in 'aeiou')
print(count)
```

```python
# 2. Write a program to check if two strings are anagrams.
a = 'asdavdd'
b = 'davddas'
a_list = sorted(a.lower())
b_list = sorted(b.lower())

if a_list == b_list:
    print('The Sentences are Anagrams')
else: print('The Sentences are not Anagrams\n')

# Using Counter
from collections import Counter
a_counter = Counter(a.lower())
b_counter = Counter(b.lower())
if a_counter == b_counter:
    print('The Sentences are Anagrams')
else: print('The Sentences are not Anagrams')
```

✅ **Practice Task:**
1. Print a hollow diamond pattern.
2. Print a multiplication table (1 to 10) using nested loops.

```python
# 1. Print a hollow diamond pattern.
for i in range(1,n+1):
    print(' '*(n-i),end=' ')
    if i==1:
        print('* ')
    elif i==n:
        print('* '*n)
    else:
        print('*'+' '*(2*i-3)+'*')
```

```python
# 2. Print a multiplication table (1 to 10) using nested loops
for i in range(1,11):
    for j in range(1,11):
        print(f"{i*j:4}",end=' ')
    print()
```
