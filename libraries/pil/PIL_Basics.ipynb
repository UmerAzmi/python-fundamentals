# 📸 PIL (Pillow) Basics

### 🔹 What is Pillow?
Pillow (PIL fork) is a Python library for working with images. It helps us:
- Open and save images in many formats (JPEG, PNG, etc.)  
- Show images  
- Resize, rotate, crop, and transform images  
- Convert images to grayscale, black & white, RGB, RGBA  
- Apply flips, brightness, contrast  
- Perform simple editing operations  


## 1. Opening and Showing Images

```python
from PIL import Image
```

```python
# Open an image
img = Image.open('my.jpg')
print(img.size)
img.show()  # Opens in default image viewer
```

## 2. Resizing Images
- `resize((w, h))` → force resize
- `thumbnail((w, h))` → resize but **keep aspect ratio**

```python
# Force resize
img.resize((2000, 2000)).show()

# Keep aspect ratio
img.thumbnail((500, 500))
img.show()
```

## 3. Rotating Images
- `rotate(angle)` rotates counter-clockwise.
- Use `expand=True` to avoid cutting parts.

```python
# Simple rotation
img.rotate(90).show()

# Rotate with expand
img.rotate(45, expand=True).show()
```

## 4. Converting Color Modes
- `'L'` → Grayscale
- `'1'` → Pure black & white
- `'RGB'` → Normal color
- `'RGBA'` → RGB + Alpha (transparency)

```python
grey = img.convert('L')   # Grayscale
grey.show()

img.convert('1').show()   # Black & White
img.convert('RGB').show() # RGB
img.convert('RGBA').show()# RGB + Transparency
```

## 5. Cropping Images
- `crop((left, top, right, bottom))`
- Cuts out a rectangle from the image.

```python
cropped = img.crop((20, 1250, 2500, 3900))
cropped.show()
```

## 6. Flipping Images
- `ImageOps.mirror()` → horizontal flip
- `ImageOps.flip()` → vertical flip

```python
from PIL import ImageOps

ImageOps.mirror(img).show()   # Horizontal flip
ImageOps.flip(img).show()     # Vertical flip
```

## 7. Image Enhancements
- We can adjust brightness, contrast, sharpness, etc.

```python
from PIL import ImageEnhance

# Brightness
enhancer = ImageEnhance.Brightness(img)
enhancer.enhance(1.5).show()   # 1.5x brighter

# Contrast
contrast = ImageEnhance.Contrast(img)
contrast.enhance(2).show()     # Double contrast
```

## 8. Saving Images
- `save(filename)` saves in a new format.

```python
grey = img.convert('1')
grey.save('BW.jpg')   # Save grayscale as BW.jpg
```

# ✅ Summary
With Pillow you can:
- **Open, show, save** images  
- **Resize** (fixed size or keep ratio)  
- **Rotate** (normal or expanded)  
- **Convert color modes** (grayscale, RGB, etc.)  
- **Crop** regions  
- **Flip** horizontally/vertically  
- **Enhance** brightness/contrast  

👉 This covers **most real-world needs** for beginner to intermediate projects.  
For advanced computer vision, you can later explore **OpenCV**.
