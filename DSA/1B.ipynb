# 242. Valid Anagram  
Given two strings `s` and `t`, return **true** if `t` is an anagram of `s`, and **false** otherwise.

### Example
Input: `s = "anagram"`, `t = "nagaram"`  
Output: `true`

```
def isAnagram(s: str, t: str) -> bool:
    return sorted(s) == sorted(t)

# Example
print(isAnagram("anagram", "nagaram"))  # True
```

**Explanation:**  
We sort both strings and compare them. If both sorted strings are identical, then they contain the same characters in the same frequency, meaning one is an anagram of the other.

# 387. First Unique Character in a String  
Given a string `s`, find the first non-repeating character in it and return its index. If it does not exist, return -1.

### Example
Input: `s = "leetcode"`  
Output: `0`

```
from collections import Counter

def firstUniqChar(s: str) -> int:
    freq = Counter(s)
    for i, ch in enumerate(s):
        if freq[ch] == 1:
            return i
    return -1

# Example
print(firstUniqChar("leetcode"))  # 0
```

**Explanation:**  
We use a frequency counter to count how many times each character appears. Then we loop through the string again and return the index of the first character that has a frequency of 1.

# 557. Reverse Words in a String III  
Given a string `s`, reverse the order of characters in each word while still preserving whitespace and the initial word order.

### Example
Input: `s = "Let's take LeetCode contest"`  
Output: `"s'teL ekat edoCteeL tsetnoc"`

```
def reverseWords(s: str) -> str:
    words = s.split()
    reverse_words = [w[::-1] for w in words]
    return ' '.join(reverse_words)

# Example
print(reverseWords("Let's take LeetCode contest"))
```

**Explanation:**  
We split the sentence into words, reverse each word individually, and then join them back together with spaces.

# 392. Is Subsequence  
Given two strings `s` and `t`, return **true** if `s` is a subsequence of `t`, otherwise return **false**.

### Example
Input: `s = "abc"`, `t = "ahbgdc"`  
Output: `true`

```
def isSubsequence(s: str, t: str) -> bool:
    i = 0
    for j in t:
        if i < len(s) and s[i] == j:
            i += 1
    return i == len(s)

# Example
print(isSubsequence("abc", "ahbgdc"))  # True
```

**Explanation:**  
We iterate through string `t` and check if characters from `s` appear in order. If all characters of `s` are matched in order inside `t`, then `s` is a subsequence of `t`.

# 412. Fizz Buzz  
Given an integer `n`, return a string array where:
- `"FizzBuzz"` if divisible by 3 and 5  
- `"Fizz"` if divisible by 3  
- `"Buzz"` if divisible by 5  
- Otherwise, the number itself

### Example
Input: `n = 3`  
Output: `["1","2","Fizz"]`

```
def fizzBuzz(n: int):
    s = []
    for i in range(1, n+1):
        if i % 3 == 0 and i % 5 == 0:
            s.append("FizzBuzz")
        elif i % 3 == 0:
            s.append("Fizz")
        elif i % 5 == 0:
            s.append("Buzz")
        else:
            s.append(str(i))
    return s

# Example
print(fizzBuzz(3))  # ["1", "2", "Fizz"]
```

**Explanation:**  
We loop through numbers from 1 to `n`. For each number, we check divisibility by 3 and 5, and append the correct string according to the rules.
