## 1. Two Sum

Given an array of integers `nums` and an integer `target`, return indices of the two numbers such that they add up to `target`.

You may assume that each input would have exactly one solution, and you may not use the same element twice.

You can return the answer in any order.

**Example:**  
Input: nums = [2,7,11,15], target = 9  
Output: [0,1]  
Explanation: Because nums[0] + nums[1] == 9, we return [0, 1].

```python
def two_sum(nums, target):
    for i in range(len(nums)):
        for j in range(i+1, len(nums)):
            if nums[j] == target - nums[i]:
                return [i, j]
    return 'Numbers not found'
two_sum([2,7,11,15],9)
```

**Explanation:**

We loop through the array with two nested loops:  
- The outer loop picks one number.  
- The inner loop checks if there exists another number such that together they equal the target.  
If found, we return their indices.  
This guarantees finding the required pair since the problem ensures one solution exists.

## 26. Remove Duplicates from Sorted Array

Given a sorted array `nums`, remove the duplicates in-place such that each unique element appears only once.  
The relative order of the elements should be kept the same.  
Return the number of unique elements.  

**Example:**  
Input: nums = [1,1,2]  
Output: 2, nums = [1,2,_]

```python
def removeDuplicates(nums):
    j = 1
    for i in range(1,len(nums)):
        if nums[i] != nums[i-1]:
            nums[j] = nums[i]
            j += 1
    print(nums)
    return j

removeDuplicates([1,2,2,5,3,2,4,1])
```

**Explanation:**

We maintain a pointer `j` which represents the position to insert the next unique element.  
- Iterate through the array starting from index 1.  
- If the current element is different from the previous, place it at index `j` and move `j` forward.  
- At the end, `j` is the count of unique elements.

## 35. Search Insert Position

Given a sorted array of distinct integers and a target value, return the index if the target is found.  
If not, return the index where it would be if inserted in order.

**Example:**  
Input: nums = [1,3,5,6], target = 5  
Output: 2

```python
def searchInsert(nums, target):
    for i in range(len(nums)):
        if nums[i] == target:
            return i
        elif nums[i] > target:
            return i
    return len(nums)

searchInsert([1,2,3,4,6],5)
```

**Explanation:**

We iterate through the list:  
- If an element equals the target, return its index.  
- If an element is greater than the target, return that index (target should be inserted here).  
- If we finish the loop, the target is greater than all elements, so we return the array’s length (insert at the end).

## 217. Contains Duplicate

Given an integer array `nums`, return **True** if any value appears at least twice in the array, and **False** if every element is distinct.

**Example:**  
Input: nums = [1,2,3,1]  
Output: true

```python
def containsDuplicate(nums):
    seen = set()
    for num in nums:
        if num in seen:
            return True
        seen.add(num)
    return False

containsDuplicate([1,2,3,5,4,6])
```

**Explanation:**

We use a set to track seen numbers:  
- Iterate through each number.  
- If it’s already in the set, return True (duplicate found).  
- Otherwise, add it to the set.  
If no duplicates are found, return False.

## 53. Maximum Subarray

Given an integer array `nums`, find the subarray with the largest sum, and return its sum.

**Example:**  
Input: nums = [-2,1,-3,4,-1,2,1,-5,4]  
Output: 6  
Explanation: The subarray [4,-1,2,1] has the largest sum 6.

```python
def maxSubArray(nums):
    current_sum = nums[0]
    max_sum = nums[0]

    for i in range(1,len(nums)):
        if current_sum < 0:
            current_sum = 0
        current_sum += nums[i]
        max_sum = max(max_sum, current_sum)

    return max_sum

maxSubArray([-2,1,-3,4,-1,2,1,-5,4])
```

**Explanation:**

We use a running sum approach:  
- Start with the first element as the current sum and max sum.  
- For each element, if the running sum becomes negative, reset it to 0 (start a new subarray).  
- Add the current number to the running sum.  
- Update max sum whenever the running sum exceeds it.  
This ensures we always capture the subarray with the maximum total.
