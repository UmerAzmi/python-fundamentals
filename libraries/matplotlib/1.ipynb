# Matplotlib Basics 📊

Matplotlib is a popular **data visualization library** in Python.  
It is used to create plots, charts, and graphs to make data easier to understand.

---

### Why use Matplotlib?
- Helps **visualize patterns** in data.
- Makes it easier to explain results.
- Works well with **NumPy & Pandas**.

---


## 1. Line Plot 📈

A **line plot** connects data points with a line.  
Useful for showing **trends over time** or continuous data.

```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4, 5]     # X-axis values
y = [2, 4, 6, 8, 10]    # Y-axis values

plt.plot(x, y)          # Creates a line plot (connects points with a line)
plt.title("Simple Line Plot")  # Adds a title
plt.xlabel("X Axis")           # Labels the X-axis
plt.ylabel("Y Axis")           # Labels the Y-axis
plt.show()                     # Displays the graph
```

## 2. Bar Chart 🍎

A **bar chart** represents data with rectangular bars.  
Useful for comparing **categories**.

```python
fruits = ['Apple', 'Mango', 'Banana']
sales = [7, 6, 12]

plt.bar(fruits, sales,color='red',edgecolor='grey')
plt.xlabel('Fruits')
plt.ylabel('Sales')
plt.title('Sales of Fruits')
plt.show()
```

## 3. Histogram 📊

A **histogram** groups numbers into **intervals (bins)** and shows how many values fall in each range.

Example explanation:
- `bins=5` → splits ages into 5 ranges (e.g., 15–18, 18–21, etc.).  
- `color='darkblue'` → fills bars with color.  
- `edgecolor='black'` → adds a border to bars for clarity.

```python
ages = [15, 18, 21, 21, 24, 30]

plt.hist(ages, bins=5, color='darkblue', edgecolor='black')
plt.title("Age Distribution")
plt.xlabel("Age Ranges")
plt.ylabel("Count")
plt.show()
```

## 4. Scatter Plot 🔵

A **scatter plot** uses points to show relationships between two variables.  
Good for seeing **correlation** (how one variable affects another).

```python
x = [5,7,8,7,6,9,5,6,7,8]
y = [99,86,87,88,100,86,103,87,94,78]

plt.scatter(x, y)
plt.title("Scatter Example")
plt.show()
```

## 5. Multiple Lines 📉📈

We can plot more than one line on the same graph.  
Use **labels** to differentiate, and `plt.legend()` to show them.

```python
x = [1,2,3,4,5]
y1 = [1,4,9,16,25]   # Squares
y2 = [1,2,3,4,5]     # Linear

plt.plot(x,y1, label="Squares",linewidth=3)
plt.plot(x,y2, label="Linear")
plt.legend()
plt.show()
```

## 6. Styled Line ✨

You can change **color, line style, and markers**.

- `color='red'` → line color.
- `linestyle='--'` → dashed line.  
- `marker='o'` → circle markers.  
- `plt.grid(True)` → adds gridlines.  
- `plt.savefig("mygraph.png")` → saves the graph as an image.

```python
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.plot(x, y, color='red', linestyle='--', marker='o',linewidth=3)
plt.title("Styled Line")
plt.grid(True)
plt.show()
plt.savefig("mygraph.png")
```

## 📌 Matplotlib Structure

- **matplotlib** → the full library (has many submodules).

- `**matplotlib.pyplot**` → Provides simple functions like `plot()`, `bar()`, `hist()`, `scatter()`, etc.
            Works like MATLAB’s plotting style (state-based, easy to use).

- `**matplotlib.figure / axes**` →  Used for **object-oriented plotting**.
            More advanced, usually applied in big or complex projects where multiple plots and fine control are required.

- `**matplotlib.animation**` → Used to create **animations** with plots.

- `**matplotlib.colors**` →  For advanced **color maps** and styling.

- `**matplotlib.image**` →  For working with and displaying **images**.
