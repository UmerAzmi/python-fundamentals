# Python Basics: Data Types, Variables, Operators, and Conditions

### 1. Data Types
- **int** â†’ whole numbers (e.g., `5`)
- **str** â†’ strings or text (e.g., `'Umer'`, `'123'`)
- **bool** â†’ Boolean values (True/False)
- **float** â†’ decimal numbers (e.g., `0.23`, `1.53`)

```python
# Convert Data types
string = '24'
print(string, type(string))   # Initially a string

integer = int(string)         # Converted to integer
print(integer, type(integer))
```

### 2. Variables
Variables are used to store values. They can have letters, numbers, and underscores but should not start with a number.

```python
# Example of variable naming
Name1_First1st = 'Umer'
print(Name1_First1st)
```

### 3. Operators
Arithmetic operators:
- `+` â†’ addition
- `-` â†’ subtraction
- `/` â†’ division (returns float)
- `//` â†’ integer division (no decimals)
- `%` â†’ modulus (remainder)
- `*` â†’ multiplication
- `**` â†’ exponentiation (power)


```python
print('Enter a number:')
num_base = input()
print('Enter a number that would be raised to')
num_expo = input()

# Convert input to int and calculate power
print(int(num_base) ** int(num_expo))
```

### 4. Conditional Operators
- `<` â†’ less than
- `>` â†’ greater than
- `==` â†’ equal to
- `!=` â†’ not equal to


```python
print('Umer' != 'Azmi')   # True
print('Umer' == 'Azmi')   # False
print(2 > 3)              # False
print(2 < 3)              # True
```

### 5. Logical Operators
- `and` â†’ True if both conditions are true
- `or` â†’ True if at least one condition is true
- `not` â†’ reverses the result


```python
y = 4
x = 3

print(y == x and x + y == 7)    # False and True -> False
print(y == x or x + y == 7)     # False or True -> True
print(not(y == x and x + y == 7)) # not(False) -> True
```

### 6. Functions â†’ ()
- `input()` â†’ allows user input
- `len()` â†’ gives the length of a string, list, or other collections


```python
# Example of len()
Name1_First1st = "Umer"
print(len(Name1_First1st))
```
