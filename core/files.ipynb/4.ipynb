## 1) Functions

- Functions group reusable code into a named block.
- Use `return` to send a value back to the caller.
- Parameters can have default values.


```python
# Simple function that returns the square of x
def squareof(x):
    # Using '**' operator to raise to power 2
    return x**2

# Example usage with input() (keeps original interactive style)
# Note: In notebooks, input() will prompt the user when the cell runs.
number = int(input('Enter a number: '))         # get user input, convert to int
print('Square is: ', squareof(number))          # call function and print result

# You can also call the function inline
print('Square is: ', squareof(int(input('Enter another number: '))))
```

```python

# Multiple parameters and default values
print('\nMultiple Parameters: ')
def multipleParameters(x, y='umer'):   # y has a default value 'umer'
    # x is required, y is optional (defaults to 'umer' if not provided)
    print('x:', x)
    print('y:', y)
    if y == 'umer' or y=='':
        print('Yo Umer')   # shows how to check the value of y
    else:
        print('Who')

# Call the function: user supplies a name (strip spaces and lower-case it)
multipleParameters(4, input('Enter a name: ').strip().lower())
```

## 2) Reading Text Files (safe and explained)

- It's good to check if a file exists before trying to open it.
- Use `with open(...) as f:` to ensure the file is closed properly.


```python
import os

FNAME = 'testfile.txt'

# If the file doesn't exist, create a small sample so this notebook cell runs without error
if not os.path.exists(FNAME):
    with open(FNAME, 'w') as _f:
        _f.write("This is a sample testfile.\nLine two.\nLine three.")

# Classic open/read (not using 'with' - shown for learning)
file = open(FNAME, 'r')   # 'r' mode = read
print('\nReading in Text Files (classic open/read):\n', file.read(), '\n')
file.seek(0)              # move cursor back to start of file

# readlines() returns a list of lines (each line includes the trailing newline '\n' if present)
f = file.readlines()
print('readlines() output:', f)

# Clean lines: remove trailing newline characters
newlist = []
for line in f:
    if line.endswith('\n'):
        # slice to remove the last character (newline)
        newlist.append(line[:-1])
    else:
        newlist.append(line)
print('Cleaned lines (method 1):', newlist)

# Alternative easier method: strip() removes leading/trailing whitespace including '\n'
newlist = []
for lins in f:
    newlist.append(lins.strip())
print('Cleaned lines (method 2 using strip):', newlist)

file.close()  # close the file when done (since we opened without 'with')
```

## 3) Writing Text Files and Context Managers

- Use `open(..., 'w')` to write (overwrites file).
- Prefer `with open(...) as f:` which closes automatically.


```python
# Writing to a file using manual open/close
file = open('testfile1.txt', 'w')   # 'w' = write (creates file or truncates it)
file.write('Umer Azmi\n')           # write a line
file.write("I'm learning how to write to a file\n")  # write another line
file.close()                        # must close when not using context manager

# Reading it back to show content
with open('testfile1.txt', 'r') as f:
    print('\nContents of testfile1.txt (after manual write):')
    print(f.read())

# Writing and reading using a context manager (preferred)
with open('testfile1.txt', 'w+') as file:
    # 'w+' opens for reading and writing and truncates file.
    file.write('Umer Azmi\n')
    file.write("I'm learning how to write to a file using context manager\n")
    file.seek(0)         # go back to the start so we can read what we just wrote
    print("Contents of testfile1.txt (using with):")
    print(file.read())
# file is closed automatically when leaving the 'with' block
```
