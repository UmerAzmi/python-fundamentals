
# Generators and Exception Handling in Python

In this notebook, we will learn about:
- **Generators**
- **Generator expressions**
- **Custom exceptions**
- **Exception handling with try-except**



## 1. Generators

- A **generator** is a function that uses the `yield` keyword instead of `return`.  
- It produces a sequence of values one at a time, instead of returning them all at once.  
- Generators are memory efficient.

Let's look at some examples:


```python

def my_gen():
    yield 'Hi'
    def another():
        variable = 'Umer Azmi'
        return f"2: {variable}"
    yield another()
    yield 'My King'

gen = my_gen()
print(next(gen))
print(next(gen))
print(next(gen))
```


### Example: Counting using a generator


```python

def count_up_to(a):
    i = 1
    while i <= a:
        yield i
        i += 1

for num in count_up_to(5):
    print(num)
```


## 2. Generator Expressions

- Similar to list comprehensions but with `()` instead of `[]`.  
- They are memory efficient because they generate values one at a time.

Example:


```python

# List comprehension (stores all values in memory)
square = [x**x for x in range(10)]
print(square)

# Generator expression (produces values one by one)
square_gen = (x**x for x in range(10))
print(next(square_gen))
print(next(square_gen))
print(next(square_gen))

gen = (x for x in range(3))
print(next(gen))  # 0
print(next(gen))  # 1
print(next(gen))  # 2
```


## 3. Exception Handling

- **Exceptions** are errors that can occur during runtime.  
- We use `try-except` blocks to handle exceptions and prevent programs from crashing.  
- We can also create **custom exceptions** by defining our own exception classes.

Example:


```python

class NegativeNumberError(Exception):  # Custom Error
    pass

def square_root(x):
    if x < 0:
        raise NegativeNumberError("Can't take square root of negative number")
    return x ** 0.5

try:
    print(square_root(-9))
except NegativeNumberError as e:
    print("Custom Error Caught:", e)
else:
    print("Square root calculated successfully")
finally:
    print("End of operation")

print('\n')
try:
    print(square_root(2))
except (ValueError, TypeError) as e:
    print("Error Caught:", type(e).__name__, e)
else:
    print("Square root calculated successfully")
finally:
    print("End of operation")
```

**`try`** → "Attempt to run this code. If everything goes smoothly, we'll continue normally."

**`except`** → "If the `try` block hits an error/exception, jump here instead of crashing. This handles the problem gracefully."

**`else`** → "If the `try` block completed successfully with **no errors**, then run this bonus code. This is for code that should only run when there are no exceptions."

**`finally`** → "No matter what happened in `try`, `except`, or `else` — whether there was an error or not — this code **will always run**. Useful for cleanup tasks like closing files."
