
# 1. Arrays / Lists  
Lists in Python are like arrays. You can store multiple values in a single variable.

**Key Features:**
- Ordered collection of items  
- Mutable (can be changed)  
- Can store mixed data types  


```python

# Creating a list
numbers = [1, 2, 3, 4, 5]

print("Accessing elements:")
print(numbers[0])   # First element -> 1
print(numbers[-1])  # Last element -> 5

print("\nUpdating elements:")
numbers[2] = 10  # Update 3rd element (index 2)
print(numbers)   # Output: [1, 2, 10, 4, 5]
```

```python
print("\nLooping through list:")
for num in numbers:
    print(num)
```

```python
print("Reverse a list")
nums = [10, 20, 30, 40]
print("Original:", nums)
print("Reversed:", nums[::-1])
nums.reverse()
print("Original:", nums)
```

```python
# Find max/min element manually and using built-ins
nums = [1,5,3,7,9,5,9,2]

min_num = nums[0]
max_num = nums[0]

for i in nums:
    if i < min_num:
        min_num = i
    if i > max_num:
        max_num = i
print(f'Minimum: {min_num} \nMaximum: {max_num}')

# Directly using min and max
print(f'\nMinimum: {min(nums)} \nMaximum: {max(nums)}')
```

```python
# Find second largest element
nums = [1,5,3,7,9,5,9,2]
first = second = float('-inf')
for n in nums:
    if n > first:
        second = first
        first = n
    elif n > second and n != first:
        second = n
print(f'Largest Element: {first} \nSecond Largest Element: {second}')
```

```python
# Remove duplicates
nums = [1,5,3,7,9,5,9,2]
unique = []
for num in nums:
    if num not in unique:
        unique.append(num)
unique.sort()
print(f'\nNums: {nums}\nUnique list: ', unique)

# Sets automatically remove duplicates
unique = set(unique)
print('Unique Sorted list: ', sorted(unique))
```

```python
# Rotate / Shift array
nums = [1,5,3,7,9,5,9,2]
k = 3
k = k % len(nums)
print(f'Rotated List: {nums[-k:]+nums[:-k]}')
```

# 2. Strings  
Strings are sequences of characters and can be treated like lists.  


```python
text = "Python"
print("\nString examples:")
print(text[0])    # First character -> P
print(text[1:4])  # Slice -> 'yth'
```

```python

print("\nReverse string:")
print(text[::-1])  # Output: nohtyP
```

```python

# Check palindrome (word that reads the same forward and backward)
while True:
    word = 'a'   # Replace with input("Enter a word: ")
    print(word)
    if word == word[::-1]:
        print('The word {wor} is a palindrome'.format(wor=word))
        break
    else:
        print(f'The word {word} is not a palindrome')
```

```python
# Check palindrome with special characters removed
import re
a = "A man, a plan, a canal: Panama"

def palin(s):
    s = re.sub(r'[^a-zA-Z0-9]', '', s).lower()
    return s == s[::-1]

print(palin(a))
```

# 3. Star & Number Patterns  
Patterns help in understanding loops and nested loops.  


```python
n = 5
```

```python
print("\nSquare pattern (3x3):")
for i in range(3):
    for j in range(3):
        print("*", end=" ")
    print()

print('\n')
for i in range(3):
    print('* '*3)
```

```python
print("\nRight Triangle pattern:")
for i in range(1, n+1):
    print("* " * i)
```

```python

print("\nPyramid pattern:")
for i in range(1, n+1):
    print(" "*(n-i) + "* "*i)
```

```python
# Inversed/Right Angled
print('\nInversed/Right Angled')
for i in range(n, 0, -1):
    print('* ' * i)
```

```python
# Increasing/left Angled
print('\nIncreasing/left Angled')
for i in range(1, n):
    print(' ' * (n - i) + '*' * i)
```

```python
# Diamond Pattern
print('\nDiamond Pattern')
for i in range(1, n):
    print(' ' * (n - i) + '* ' * i)
for i in range(n, 0, -1):
    print(' ' * (n - i) + '* ' * i)
```

```python
# Pyramid Pattern
print('\nPyramid Pattern')
for i in range(1,n):
    print(' '*(n-i)+'*'*(2*i-1))
```

```python
# Hollow Triangle
print('\nHollow Triangle')
for i in range(1,n+1):
    print(' '*(n-i), end='')
    if i==1:
        print('*')
    elif i==n:
        print('* '*n)
    else:
        print('*'+' '*(2*i-3)+'*')
```

```python
# Number Pyramid
print('\nNumber Pyramid')
for i in range(1,n):
    print(' '*(n-i),end='')
    for j in range(1,i+1):
        print(j,end='')
    for j in range(i-1,0,-1):
        print(j,end='')
    print()
```
