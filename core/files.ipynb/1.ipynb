# Python Basics: Data Types, Variables, Operators, and Conditions

### 1. Data Types
- **int** → whole numbers (e.g., `5`)
- **str** → strings or text (e.g., `'Umer'`, `'123'`)
- **bool** → Boolean values (True/False)
- **float** → decimal numbers (e.g., `0.23`, `1.53`)

```
# Convert Data types
string = '24'
print(string, type(string))   # Initially a string

integer = int(string)         # Converted to integer
print(integer, type(integer))
```

### 2. Variables
Variables are used to store values. They can have letters, numbers, and underscores but should not start with a number.

```
# Example of variable naming
Name1_First1st = 'Umer'
print(Name1_First1st)
```

### 3. Operators
Arithmetic operators:
- `+` → addition
- `-` → subtraction
- `/` → division (returns float)
- `//` → integer division (no decimals)
- `%` → modulus (remainder)
- `*` → multiplication
- `**` → exponentiation (power)


```
print('Enter a number:')
num_base = input()
print('Enter a number that would be raised to')
num_expo = input()

# Convert input to int and calculate power
print(int(num_base) ** int(num_expo))
```

### 4. Conditional Operators
- `<` → less than
- `>` → greater than
- `==` → equal to
- `!=` → not equal to


```
print('Umer' != 'Azmi')   # True
print('Umer' == 'Azmi')   # False
print(2 > 3)              # False
print(2 < 3)              # True
```

### 5. Logical Operators
- `and` → True if both conditions are true
- `or` → True if at least one condition is true
- `not` → reverses the result


```
y = 4
x = 3

print(y == x and x + y == 7)    # False and True -> False
print(y == x or x + y == 7)     # False or True -> True
print(not(y == x and x + y == 7)) # not(False) -> True
```

### 6. Functions → ()
- `input()` → allows user input
- `len()` → gives the length of a string, list, or other collections


```
# Example of len()
Name1_First1st = "Umer"
print(len(Name1_First1st))
```
