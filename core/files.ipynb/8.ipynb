
# Class Variables, Class Methods, and Static Methods in Python

In this notebook, we will learn about:
- **Class Variables**
- **Class Methods**
- **Static Methods**
- The difference between **public** and **private** variables/methods.



## 1. Class Variables

- A **class variable** is shared by all objects of the class.  
- It is defined inside the class but **outside** any methods.

In the example below, `dogs` is a class variable that stores all dog names.


````python

class Dogs:
    dogs = []   # Class Variable
    
    def __init__(self, name, age=None):
        self.name = name
        self.age = age
        self.dogs.append(self.name)  # Appending name to the class variable list
````


## 2. Class Methods

- A **class method** is defined using the `@classmethod` decorator.  
- It takes `cls` (the class itself) as the first parameter instead of `self`.  
- It can access and modify **class variables**.

In this example:
- `total()` returns the number of dogs.  
- `nameAge()` creates a new `Dogs` object from a string like `"Tim-20"`.


````python

class Dogs:
    dogs = []   # Class Variable
    
    def __init__(self, name, age=None):
        self.name = name
        self.age = age
        self.dogs.append(self.name)
    
    @classmethod
    def total(cls):
        return len(cls.dogs)
    
    @classmethod
    def nameAge(cls, data):
        names, age = data.split('-')
        return cls(names, int(age))
````


## 3. Static Methods

- A **static method** is defined using the `@staticmethod` decorator.  
- It does **not** take `self` or `cls` as the first parameter.  
- It behaves like a normal function but belongs to the class for logical grouping.

In this example, `dawg()` is a static method that simply prints a message.


````python

class Dogs:
    dogs = []   # Class Variable
    
    def __init__(self, name, age=None):
        self.name = name
        self.age = age
        self.dogs.append(self.name)
    
    @classmethod
    def total(cls):
        return len(cls.dogs)
    
    @classmethod
    def nameAge(cls, data):
        names, age = data.split('-')
        return cls(names, int(age))
    
    def __str__(self):
        return f"Name:{self.name}, Age:{self.age}"
    
    @staticmethod
    def dawg(name, age):
        return f"Ssup Dawg, I'm {name} and I'm {age} years old SON!!"
````


## 4. Using the `Dogs` Class

Let's create some dogs and test class variables, class methods, and static methods.


````python

a, b, c, d = Dogs('A'), Dogs('B'), Dogs('C'), Dogs('D')

# Accessing the class variable
print('Class Variable Call:', Dogs.dogs)

# Using class method
print('Count:', Dogs.total())

# Using class method to create a new object
aDogs = Dogs.nameAge("Tim-20")
print(aDogs)

# Using static method
print(Dogs.dawg('Umer', 22))

# Checking updated class variable
print(Dogs.dogs)
````


## 5. Public and Private Variables/Methods

- In Python, we use an **underscore (`_`)** before a variable or method name to indicate that it is **private**.  
- Private attributes/methods are not truly hidden, but it's a convention to say "this should not be accessed directly".

Example:
```python
class Example:
    def __init__(self):
        self.public_var = "I am public"
        self._private_var = "I am private"
```
