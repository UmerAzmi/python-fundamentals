# 👁️ OpenCV Basics

### 🔹 What is OpenCV?
OpenCV (Open Source Computer Vision Library) is a powerful Python library for **image and video processing**.

With OpenCV, you can:
- Open and save images in different formats  
- Display images and videos  
- Resize, crop, and rotate images  
- Convert images to grayscale or other color formats  
- Draw shapes and text on images  
- Detect faces and objects  
- Work with webcam or video streams in real time  

👉 Learning these basics will allow you to build projects like **face detection, QR code scanning, filters, motion detection**, etc.


## 1. Reading and Displaying Images
OpenCV loads images using `cv2.imread()` and shows them using `cv2.imshow()`. Always use `cv2.waitKey(0)` to keep the window open until a key is pressed.

```
import cv2

# Read image (by default it loads in BGR format)
img = cv2.imread('my.jpg')

# Show image in a window
cv2.imshow('Original Image', img)
cv2.waitKey(0)  # Wait until any key is pressed
cv2.destroyAllWindows()
```

## 2. Converting to Grayscale
Color images are made of 3 channels (Blue, Green, Red). For many tasks (like face detection), grayscale is enough and faster to process.

```
# Convert image to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

cv2.imshow('Grayscale Image', gray)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## 3. Resizing and Cropping
- **Resize**: Use `cv2.resize()` to change dimensions.
- **Crop**: Select part of the image using slicing `[y1:y2, x1:x2]`.

```
# Resize image to 400x400
resized = cv2.resize(img, (400, 400))
cv2.imshow('Resized Image', resized)

# Crop region of interest (y from 50 to 300, x from 100 to 400)
cropped = img[50:300, 100:400]
cv2.imshow('Cropped Image', cropped)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

## 4. Drawing Shapes and Text
We can draw directly on images using OpenCV functions:
- `cv2.line()` → draw line
- `cv2.rectangle()` → draw rectangle
- `cv2.circle()` → draw circle
- `cv2.putText()` → add text

```
# Draw a green line
cv2.line(img, (50, 50), (200, 200), (0, 255, 0), 3)

# Draw a blue rectangle
cv2.rectangle(img, (100, 100), (300, 300), (255, 0, 0), 2)

# Draw a red circle
cv2.circle(img, (250, 250), 100, (0, 0, 255), 3)

# Add white text
cv2.putText(img, "Hello OpenCV", (50, 400), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 255), 2)

cv2.imshow('Drawing Example', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## 5. Working with Video
We can capture video from a webcam or file using `cv2.VideoCapture()`. This reads frames in a loop and displays them in real-time.

```
# Capture video from default webcam (0)
cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()  # Read one frame
    if not ret:
        break

    cv2.imshow('Webcam', frame)

    # Press 'q' to quit
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

## 6. Face Detection Project
OpenCV provides **Haar Cascade Classifiers** (pre-trained models) for detecting faces, eyes, etc.

Steps:
1. Convert image to grayscale
2. Load face detection model
3. Detect faces with `detectMultiScale()`
4. Draw rectangles around detected faces

```
# Load the pre-trained face detection model
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

# Read input image
img = cv2.imread('face.jpg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Detect faces (scaleFactor=1.3, minNeighbors=5)
faces = face_cascade.detectMultiScale(gray, 1.3, 5)

# Draw rectangles around detected faces
for (x, y, w, h) in faces:
    cv2.rectangle(img, (x, y), (x+w, y+h), (255, 0, 0), 2)

cv2.imshow('Detected Faces', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
