# Modular Programming
Modular programming allows us to split code into different files (modules) and import them when needed.

- `math` is a built-in Python module with many mathematical functions and constants.
- `myModule` is our custom module (saved as **myModule.py** in the same folder).
    - `myFunc(x)` → returns `x+5`
    - `anotherfunc(x)` → returns `x//5`


```
import math
import myModule

print(math.pi)                 # Prints the value of π (pi)
print(myModule.myFunc(5))      # 5 + 5 = 10
print(myModule.anotherfunc(5)) # 5 // 5 = 1
```

# Try and Except (Exception Handling)
`try` and `except` blocks help us handle runtime errors gracefully without crashing the program.


```
text = input('Username: ')
try:
    number = int(text)  # Try converting input to integer
    print(number)
except:
    print('Invalid Username')
```

## Example 1: Local variable does NOT affect global variable
If a variable inside a function has the same name as a global one, Python treats it as a **new local variable**.


```
loop = True    ## Global variable

def func(x):
    loop = 7   # Local variable (does not overwrite global 'loop')
    print('\nFirst Example Output:')
    print('Local loop:', loop)
    if x == 5:
        print(7)

func(2)
print('Global loop remains:', loop,'\n')  # Still True
```

## Example 2: Using `global` keyword
With the `global` keyword, the function will modify the global variable instead of creating a new local one.


```
loop = True    ## Global variable

def func(x):
    global loop   # Now function modifies global 'loop'
    loop = 7
    print('Inside function, loop =', loop)
    if x == 5:
        print(7)

func(2)
print('Global loop changed to:', loop)
```
