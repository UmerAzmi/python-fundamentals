# String Operations with Sorting, Regex, Reverse, Split, and Join

This topics covers some important string and list operations in Python such as:
- **sorted()**
- **re.sub()**
- **reverse()**
- **split()**
- **join()**


```python
import re
from collections import Counter

word = "Umer $ Azmi' ? 'UA"
words = ["apple", "banana", "cherry"]
print('\nWord:', word)

# sorted(): Returns a sorted list of characters (case-insensitive here)
sort = sorted(word.lower())

# re.sub(): Remove all non-alphanumeric characters using regex
sub = re.sub(r'[^a-zA-Z0-9]','',word)

# reverse(): Reverse the list of words
words.reverse()

print('Sorted: ',sort)
print('Subbed: ',sub)
print('Reversed: ',words)
```

### Explanation
- **`sorted(word.lower())`** → sorts all characters in the string alphabetically, ignoring case.
- **`re.sub(r'[^a-zA-Z0-9]', '', word)`** → removes everything except letters and digits.
- **`words.reverse()`** → reverses the list `['apple','banana','cherry']` in place.


```python
# Example: Reverse only the letters of each word in a string
s = "maam koot reh tayag ot eht ,ekal tub ,wow a pup depmuj !ni!, elbuod  secaps"
words = s.split(' ')                                   # Split string into words  ()

# Reverse each word using slicing [::-1]
reverse_words = [w[::-1] for w in words]

print('\nReversed Words:\n',reverse_words)

# Join them back into a single sentence
print(' '.join(reverse_words))
```

### Explanation
- **`s.split()`** → splits the sentence into words based on spaces.
- **`s.split()`** (No parameter)  - **`s.split(' ')`** (With a parameter)
- **`[w[::-1] for w in words]`** → reverses each word individually.
- **`' '.join(reverse_words)`** → joins the list back into a sentence separated by spaces.


# Extra Functions

This topic covers following functions:
- **.index()**
- **abs()**
- **isinstance()**
- **bin()**
- **.replace()**
- **bin()**
- **.capitalize()**
- **.title()**

```python
haystack = ['283497238987234', 'a dog', 'a cat', 'some random junk', 'a piece of hay', 'needle', 'something somebody lost a while ago']
def find_needle(haystack):
    return f'found the needle at position {haystack.index("needle")}'
find_needle(haystack)
```

### Explanation
- **`index()`**: Finds the starting position of a substring
- **`haystack.index("needle")`** → finds the starting index (position) of the substring `"needle"` within the `haystack` string. It returns `5`.

```python
def make_number_negative(number) :
    return -abs(number)     # Abs function returns absolute value -10 becomes 10, 10 remains 10.

make_number_negative(10)
```

### Explanation
- **`abs()`**: Gets the absolute (non-negative) value of a number
- **`abs(number)`** → returns the absolute (non-negative) value of `-15`, which is `15`.

```python
my_list = [1, 2, 'ab', 3, '4']
my_string = "Hello"

print(isinstance(my_list, list))
print(isinstance(my_string, int))
print(isinstance(my_list, (int, float)))

for item in my_list:
    if isinstance(item, int):
        print(item,end=' ')
```

### Explanation
- **`isinstance()`**: Checks if an object is an instance of a specific type
- **`isinstance(my_list, list)`** → checks if the `my_list` variable is an instance of the `list` type, returning `True`.
- **`isinstance(my_string, int)`** → checks if the `my_string` variable is an instance of the `list` type, returning `False`.

```python
text = 'uMEr  aZmI'
print(text.replace(' ',''))

print(bin(5))
print(bin(5+1)[2:])

print(text.capitalize())
print(text.title())
```

### Explanation
* **`bin(5)`** $\rightarrow$ A built-in function that converts an integer (like `5`) into its binary string representation, prefixed with `"0b"` (e.g., `"0b101"`).
* **`text.replace(' ', '')`** $\rightarrow$ A string method that searches for a substring (`' '`) and returns a new string where all occurrences are replaced with another (`''`).
* **`text.capitalize()`** $\rightarrow$ A string method that returns a copy of `text` with only its **first character** capitalized (and the rest lowercase).
* **`text.title()`** $\rightarrow$ A string method that returns a copy of `text` with the **first character of every word** capitalized (and the rest lowercase).
